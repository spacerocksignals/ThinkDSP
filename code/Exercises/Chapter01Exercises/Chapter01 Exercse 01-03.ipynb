```python
cos_sig = CosSignal(freq=262, amp=1.0, offset=0)
sin_sig = SinSignal(freq=330, amp=0.5, offset=0)
```

```python
mix = sin_sig + cos_sig
```

```python
mix.plot()
```

```python
wave = mix.make_wave(duration=0.5, start=0, framerate=11025)
```

```python
audio = Audio(data=wave.ys, rate=wave.framerate)
audio
```

```python
segment = wave.segment()
spectrum = segment.make_spectrum()
spectrum.plot()
```

```python
midC = SinSignal(freq=262, amp=1, offset=0)
midE = SinSignal(freq=330, amp=1, offset=0)
midG = SinSignal(freq=392, amp=1, offset=0)

midmaj = midC + midE + midG
midmaj.plot()
```
