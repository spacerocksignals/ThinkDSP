Exercise 2.3 Make a square signal at 1100 Hz and make a wave that samples
it at 10000 frames per second. If you plot the spectrum, you can see that most
of the harmonics are aliased. When you listen to the wave, can you hear the
aliased harmonics?

```python
from thinkdsp import SquareSignal

signal = SquareSignal(1100)
duration = 0.5
framerate = 10000
segment = signal.make_wave(duration=duration, framerate=framerate)
#segment.plot()
spectrum = segment.make_spectrum()
spectrum.plot()
```

```python
segment.make_audio()
```

```python
aliased = SquareSignal(8900).make_wave(duration=duration, framerate=framerate)
aliased.make_audio()
```
