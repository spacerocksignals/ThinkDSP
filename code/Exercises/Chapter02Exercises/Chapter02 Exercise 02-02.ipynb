```python
from thinkdsp import Sinusoid
from thinkdsp import normalize, unbias
import numpy as np

class SawtoothSignal(Sinusoid):
    
    def evaluate(self, ts):
        
        cycles = self.freq * ts + self.offset / np.pi / 2
        frac, _ = np.modf(cycles) #this goes from 0 to 1 at an interval with a given frequency
        ys = normalize(unbias(frac), self.amp) #centers the waveform at 0
        return ys
```

```python
sawtooth = SawtoothSignal(200)
duration =  sawtooth.period*3
wave = sawtooth.make_wave(.5, framerate=40000)
wave.make_audio()
```

```python
spectrum = wave.make_spectrum()
spectrum.plot(high=5000)
```

Harmonic structure of a sawtooth wave looks closer to that of a normal wave, with harmonics at every even and odd multiple (as opposed only the odd multiples with triangle or square waves).
